
### **5. How would you reverse a string in Python without using built-in functions?**

**Answer in simple words:**
Reversing a string means writing it from the last character to the first. Without built-in functions, we can use a loop.

✅ **Real-life examples:**

1. Imagine reading a book from the last page back to the first.
2. If your name is on a paper as **CHIRAG**, reversing gives **GARHIC** (reading backward).

✅ **Coding example:**

```python
text = "Python"
reversed_text = ""
for char in text:
    reversed_text = char + reversed_text
print(reversed_text)   # Output: nohtyP
```

### **6. Explain string immutability in Python.**

**Answer in simple words:**
Strings cannot be changed once created. If you try to modify, Python creates a new string instead.

✅ **Real-life examples:**

1. Think of writing your name on stone. If you want to change it, you need a new stone.
2. A printed book page cannot be edited—you must reprint a new page.

✅ **Coding example:**

```python
s = "Hello"
s = s + " World"   # New string created
print(s)           # Output: Hello World
```

### **7. What are the advantages of using lists in Python?**

**Answer in simple words:**
Lists can store many items together, in order, and can be changed anytime.

✅ **Real-life examples:**

1. A shopping list: you can add, remove, or change items.
2. A playlist of songs: you can rearrange, add new songs, or delete old ones.

✅ **Coding example:**

```python
fruits = ["apple", "banana", "cherry"]
fruits.append("mango")
print(fruits)   # Output: ['apple', 'banana', 'cherry', 'mango']
```

### **8. How are Python sets different from lists and tuples?**

**Answer in simple words:**

* **Lists/Tuples** → Keep order, can have duplicates.
* **Sets** → Do not keep order, and automatically remove duplicates.

✅ **Real-life examples:**

1. A **set of colors** in a paint box—no repeats.
2. A **list of students in class** may have duplicates if someone is counted twice.

✅ **Coding example:**

```python
my_list = [1, 2, 2, 3]
my_set = {1, 2, 2, 3}
print(my_list)  # [1, 2, 2, 3]
print(my_set)   # {1, 2, 3}
```

### **9. What is the difference between a shallow copy and a deep copy in lists?**

**Answer in simple words:**

* **Shallow copy** → Copies only the outer list, not inner lists.
* **Deep copy** → Copies everything, including inner lists.

✅ **Real-life examples:**

1. Shallow copy: photocopying a file cover but leaving old papers inside.
2. Deep copy: photocopying both the cover and every page inside.

✅ **Coding example:**

```python
import copy
list1 = [[1, 2], [3, 4]]

shallow = copy.copy(list1)
deep = copy.deepcopy(list1)

list1[0][0] = 9
print(shallow)  # [[9, 2], [3, 4]]
print(deep)     # [[1, 2], [3, 4]]
```

### **10. When would you use a dictionary over a list?**

**Answer in simple words:**
Use a dictionary when you want to store data as **key-value pairs** for fast lookup. Use a list for simple ordered items.

✅ **Real-life examples:**

1. Dictionary → Phonebook: {"John": 98765, "Sara": 12345}.
2. List → Shopping items: \["milk", "bread", "eggs"].

✅ **Coding example:**

```python
phonebook = {"John": 98765, "Sara": 12345}
print(phonebook["John"])   # Output: 98765
```

### **11. How do you merge two dictionaries in Python?**

**Answer in simple words:**
We can combine two dictionaries into one using `update()` or `{**dict1, **dict2}`.

✅ **Real-life examples:**

1. Merging two contact lists from two phones.
2. Combining two recipe books into one big recipe book.

✅ **Coding example:**

```python
dict1 = {"a": 1, "b": 2}
dict2 = {"c": 3, "d": 4}

merged = {**dict1, **dict2}
print(merged)   # Output: {'a': 1, 'b': 2, 'c': 3, 'd': 4}
``
